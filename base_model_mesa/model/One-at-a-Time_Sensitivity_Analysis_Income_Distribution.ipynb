```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sensitivity_analysis_path = r'../output_data/SensitivityAnalysisIncomeDistribution.csv'
sensitivity_analysis = pd.read_csv(sensitivity_analysis_path)
#sensitivity_analysis
```

First, we divide all the dataframe based on the income distribution that has been chosen for that model run. This way, we can analyse easier seperately. The different income distributions are here:

 income_distribution_base = {'Poor': [5000, 1875], 'Middle-Class': [29375, 10312], 'Rich': [87500, 18750]},
 income_distribution_plus_10 = {'Poor': [5500, 1875], 'Middle-Class': [32312.5, 10321], 'Rich': [96250, 18750]}
 income_distribution_minus_10 = {'Poor': [4500, 1875], 'Middle-Class': [26437.5, 10321], 'Rich': [78750, 18750]}
 income_distribution_plus_30 = {'Poor': [6500, 1875], 'Middle-Class': [38187.5, 10321], 'Rich': [133750, 18750]}
 income_distribution_minus_30 = {'Poor': [3500, 1875], 'Middle-Class': [20562.5, 10321], 'Rich': [61250, 18750]}
 
With this, we can analyse the different behaviours in the models depending on different income distribution, ran one at a time. 

```python
income_distribution_base = pd.DataFrame()

base = str({'Poor': [5000, 1875], 'Middle-Class': [29375, 10312], 'Rich': [87500, 18750]})

condition = sensitivity_analysis['income_distribution'] == base

income_distribution_base_df = pd.concat([sensitivity_analysis[condition], income_distribution_base], ignore_index=True)

#income_distribution_base_df
```

```python
income_distribution_plus_10 = pd.DataFrame()

plus_10 = str({'Poor': [5500, 1875], 'Middle-Class': [32312.5, 10321], 'Rich': [96250, 18750]})

condition = sensitivity_analysis['income_distribution'] == plus_10

income_distribution_plus_10_df = pd.concat([sensitivity_analysis[condition], income_distribution_plus_10], ignore_index=True)

#income_distribution_plus_10_df
```

```python
income_distribution_minus_10 = pd.DataFrame()

minus_10 = str({'Poor': [4500, 1875], 'Middle-Class': [26437.5, 10321], 'Rich': [78750, 18750]})

condition = sensitivity_analysis['income_distribution'] == minus_10

income_distribution_minus_10_df = pd.concat([sensitivity_analysis[condition], income_distribution_minus_10], ignore_index=True)

#income_distribution_minus_10_df
```

```python
income_distribution_plus_30 = pd.DataFrame()

plus_30 = str({'Poor': [6500, 1875], 'Middle-Class': [38187.5, 10321], 'Rich': [133750, 18750]})

condition = sensitivity_analysis['income_distribution'] == plus_30

income_distribution_plus_30_df = pd.concat([sensitivity_analysis[condition], income_distribution_plus_30], ignore_index=True)

#income_distribution_plus_30_df
```

```python
income_distribution_minus_30 = pd.DataFrame()

minus_30 = str({'Poor': [3500, 1875], 'Middle-Class': [20562.5, 10321], 'Rich': [61250, 18750]})

condition = sensitivity_analysis['income_distribution'] == minus_30

income_distribution_minus_30_df = pd.concat([sensitivity_analysis[condition], income_distribution_minus_30], ignore_index=True)

income_distribution_minus_30_df
```

# Poor households

```python
#First, poor households get isolated
estimated_ratio_base = income_distribution_base_df[['RunId', 'Step', 'EstimatedAverageIncomeToDamagePoorHousehold']]
estimated_ratio_plus_10 = income_distribution_plus_10_df[['RunId', 'Step', 'EstimatedAverageIncomeToDamagePoorHousehold']]
estimated_ratio_minus_10 = income_distribution_minus_10_df[['RunId', 'Step', 'EstimatedAverageIncomeToDamagePoorHousehold']]
estimated_ratio_plus_30 = income_distribution_plus_30_df[['RunId', 'Step', 'EstimatedAverageIncomeToDamagePoorHousehold']]
estimated_ratio_minus_30 = income_distribution_minus_30_df[['RunId', 'Step', 'EstimatedAverageIncomeToDamagePoorHousehold']]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Assuming 'estimated_ratio_base' is your DataFrame with columns 'RunId', 'Step', and 'EstimatedAverageIncomeToDamagePoorHousehold'
# Assuming 'estimated_ratio_plus_10' is another DataFrame with the same structure
# Assuming 'estimated_ratio_minus_10' is another DataFrame with the same structure
# Assuming 'estimated_ratio_plus_30' is another DataFrame with the same structure
# Assuming 'estimated_ratio_minus_30' is another DataFrame with the same structure

# Calculate the mean and standard deviation for each step in the first DataFrame
mean_values_base = estimated_ratio_base.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].mean()
std_values_base = estimated_ratio_base.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].std()

# Calculate the mean and standard deviation for each step in the second DataFrame
mean_values_plus_10 = estimated_ratio_plus_10.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].mean()
std_values_plus_10 = estimated_ratio_plus_10.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].std()

# Calculate the mean and standard deviation for each step in the third DataFrame
mean_values_minus_10 = estimated_ratio_minus_10.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].mean()
std_values_minus_10 = estimated_ratio_minus_10.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].std()

# Calculate the mean and standard deviation for each step in the fourth DataFrame
mean_values_plus_30 = estimated_ratio_plus_30.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].mean()
std_values_plus_30 = estimated_ratio_plus_30.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].std()

# Calculate the mean and standard deviation for each step in the fifth DataFrame
mean_values_minus_30 = estimated_ratio_minus_30.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].mean()
std_values_minus_30 = estimated_ratio_minus_30.groupby('Step')['EstimatedAverageIncomeToDamagePoorHousehold'].std()

# Plot the mean line for the first DataFrame
ax.plot(mean_values_base.index, mean_values_base, label='Mean (Base)')

# Fill between the upper and lower bounds for the first DataFrame (mean Â± 1 standard deviation)
ax.fill_between(mean_values_base.index, mean_values_base - std_values_base, mean_values_base + std_values_base, alpha=0.2, label='Variability (Base)')

# Plot the mean line for the second DataFrame
ax.plot(mean_values_plus_10.index, mean_values_plus_10, label='Mean (Plus 10)')

# Fill between the upper and lower bounds for the second DataFrame (mean Â± 1 standard deviation)
ax.fill_between(mean_values_plus_10.index, mean_values_plus_10 - std_values_plus_10, mean_values_plus_10 + std_values_plus_10, alpha=0.2, label='Variability (Plus 10)')

# Plot the mean line for the third DataFrame
ax.plot(mean_values_minus_10.index, mean_values_minus_10, label='Mean (Minus 10)')

# Fill between the upper and lower bounds for the third DataFrame (mean Â± 1 standard deviation)
ax.fill_between(mean_values_minus_10.index, mean_values_minus_10 - std_values_minus_10, mean_values_minus_10 + std_values_minus_10, alpha=0.2, label='Variability (Minus 10)')

# Plot the mean line for the fourth DataFrame
ax.plot(mean_values_plus_30.index, mean_values_plus_30, label='Mean (Plus 30)')

# Fill between the upper and lower bounds for the fourth DataFrame (mean Â± 1 standard deviation)
ax.fill_between(mean_values_plus_30.index, mean_values_plus_30 - std_values_plus_30, mean_values_plus_30 + std_values_plus_30, alpha=0.2, label='Variability (Plus 30)')

# Plot the mean line for the fifth DataFrame
ax.plot(mean_values_minus_30.index, mean_values_minus_30, label='Mean (Minus 30)')

# Fill between the upper and lower bounds for the fifth DataFrame (mean Â± 1 standard deviation)
ax.fill_between(mean_values_minus_30.index, mean_values_minus_30 - std_values_minus_30, mean_values_minus_30 + std_values_minus_30, alpha=0.2, label='Variability (Minus 30)')

# Customize labels and title
ax.set_xlabel('Step')
ax.set_ylabel('EstimatedAverageIncomeToDamagePoorHousehold')
ax.set_title('Estimated Average Income To Damage for Poor Households with five different income distributions as input')

# Show legend
ax.legend()

# Display the plot
plt.show()
```
